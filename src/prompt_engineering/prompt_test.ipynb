```python
def load_json(path):
    with open(path, "r", encoding="utf-8") as f:
        return json.load(f)

def deal_data(data_path, test_data_path):
    data = load_json(data_path)
    test_data = load_json(test_data_path)
    results = []
    for x,y in zip(data, test_data):
        results.append({
            'id': x['id'],
            'content': x['content'],
            'ground_truth': y['output'],
            'output': x['output']
        })
    with open(data_path, "w", encoding="utf-8") as f:
        json.dump(results, f, ensure_ascii=False, indent=4)
```

```python
train_data = load_json(train_data_path)
test_data = load_json(test_data_path)
```

```python
from text2vec import Word2Vec
w2v_model = Word2Vec("w2v-light-tencent-chinese")
```

```python

sentences = [item['content'] for item in train_data]
sentence_embeddings = w2v_model.encode(sentences, show_progress_bar=True, normalize_embeddings=True)
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def search_w2v(query: str, top_k: int = 1):
    # 1. 编码 query
    query_embedding = w2v_model.encode(
        [query],
        show_progress_bar=False,
        normalize_embeddings=True
    )  # shape: (1, dim)

    # 2. 计算与所有训练样本的余弦相似度
    sims = cosine_similarity(query_embedding, sentence_embeddings)[0]  # shape: (N,)

    # 3. 取相似度最高的 top_k 条
    top_idx = np.argsort(-sims)[:top_k]

    results = []
    for idx in top_idx:
        item = train_data[idx]
        results.append(
            {
                "score": float(sims[idx]),
                "content": item["content"],
                "output": item["output"],
            }
        )
    return results
```

```python
from src.RAG.get_rag_prompt import rag_selftrain_prompt
serach_prompt_data = []
for item in test_data:
    query = item['content']
    example = search_w2v(query, top_k=1)[0]
    prompt = rag_selftrain_prompt(query, example)
    serach_prompt_data.append({
        'id': item['id'],
        'content': query,
        'candidates': example,
        'output': item['output'],
        'prompt': prompt
    })
with open(search_prompt_path, "w", encoding="utf-8") as f:
    json.dump(serach_prompt_data, f, ensure_ascii=False, indent=4)

```
